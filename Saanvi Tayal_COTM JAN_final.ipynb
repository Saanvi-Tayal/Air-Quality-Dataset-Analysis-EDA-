## <b>Cleaning the dataset</b>

```python
df['NMHC_GT'].value_counts()
```

```python
# 8443 values out of 9357 are -200. This coulmn is not needed thus dropping this coulumn .
df.drop('NMHC_GT', axis=1, inplace=True)
```

```python
df.columns
```

```python
# In data we  lot of -200 which seems to be a error. Removing those values and replacing with Nan i.e null 
df.replace(to_replace= -200, value= np.NaN, inplace= True)
```

```python
# Use fillna function to fill the missing values with an estimate value
col_list = df.columns[2:13]

for i in col_list:
    df[i] = df[i].fillna(df[i].mean())
```

```python

columns = ["Nox_GT", "PT08_S3_Nox", "NO2_GT"]
oneday = df[6:28].copy()

fig =go.Figure([{'x': df["datetime"],'y': df[col],'name': col}  for col in columns], layout=go.Layout(title=go.layout.Title(text="1 Year Time Span of Air Concentrations")))    
fig.show()
```

### Extracting features of Air depending upon the level of CO concentration.

```python
df['CO_level'].unique()
CO_level_count = df['CO_level'].value_counts()
print(CO_level_count)
plt.pie(CO_level_count, labels = CO_level_count.index,colors=['orange','red','green','blanchedalmond'],radius = 2,shadow= True,autopct = "%.2f%%");
plt.show()
```

```python
#relation between CO concentration and temperature, Relative Humidity, Absolute Humidity
fig, axes = plt.subplots(1,3, figsize=(25, 5))
sns.violinplot(x='CO_level', y='T', data=df, hue="CO_level", palette='rocket', ax=axes[0])
axes[0].set_title("{} by {}".format("CO_level", "T"))

    
sns.violinplot(x="CO_level", y="CO_GT", data=df, hue="CO_level", palette='rocket', ax=axes[1])
axes[1].set_title("{} by {}".format("CO_level", "CO_GT"))

    
sns.violinplot(x="CO_level", y="RH", data=df, hue="CO_level", palette='rocket', ax=axes[2])
axes[2].set_title("{} by {}".format("CO_level", "RH"))
```
